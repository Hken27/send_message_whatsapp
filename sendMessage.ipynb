```python
import pywhatkit
import time
import pyautogui
import pandas as pd
from pynput.keyboard import Controller
```

```python


keyboard = Controller()

def send_whatsapp_image(phone_no: str, image_path: str, caption: str):
    try:
        pywhatkit.sendwhats_image(
            receiver=phone_no, 
            img_path=image_path, 
            caption=caption,
            wait_time=15,
            tab_close=True  # Close the tab after sending the image
        )
        time.sleep(4)
        pyautogui.click()
        time.sleep(4)
        
        
        pyautogui.press('enter') 
        print(f"Image sent to {phone_no} successfully!")
    except Exception as e:
        print(f"Failed to send image to {phone_no}: {str(e)}")
        
        
def send_whatsapp_message(phone_no: str, msg: str):
    try:
        pywhatkit.sendwhatmsg_instantly(
            phone_no=phone_no, 
            message=msg,
            tab_close=True  # Close the tab after sending the message
        )
        time.sleep(4)
        pyautogui.click()
        time.sleep(4)
        pyautogui.press('enter') 
        print(f"Message sent to {phone_no} successfully!")
        
    except Exception as e:
        print(f"Failed to send message to {phone_no}: {str(e)}")

 
 
def send_messages_from_excel(file_path: str, image_path: str):
    try:
        data = pd.read_excel(file_path)
        for index, row in data.iterrows():
            phone_no = row['Handphone']           
            message = row['Message']
            # send_whatsapp_message(phone_no, message)
            send_whatsapp_image(phone_no, image_path, message)
            time.sleep(15)  
    except Exception as e:
        print(f"Error reading Excel file: {str(e)}")
 
if __name__ == "__main__":
    image_path = 'optional'  # Fixed image path
    file_path = 'your_file.xlsx' 
    send_messages_from_excel(file_path, image_path)

```
